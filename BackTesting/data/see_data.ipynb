```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('btcusdt_3m_train.csv')
df = df[df['volume']!=0]
```

```python
df['ret'] = df['close'] - df['open']
df.dropna(inplace=True)
df['sign'] = np.sign(df['ret'])
print(df['sign'].value_counts())
```

```python
i = 0
j = 0
data = {"R":{}, "G":{}}
while j < df.shape[0]:
    while df.iloc[i]['sign'] == df.iloc[j]['sign']:
        j += 1
        if j == df.shape[0]:
            break
    len = j - i 
    if df.iloc[i]['sign'] == 1:
        if len not in data["G"]:
            data["G"][len] = 0
        data["G"][len] += 1
    else:
        if len not in data["R"]:
            data["R"][len] = 0
        data["R"][len] += 1
    i = j
    
print(data)

```

```python
# sort by key
data["R"] = sorted(data["R"].items(), key=lambda d:d[0], reverse=True)
data["G"] = sorted(data["G"].items(), key=lambda d:d[0], reverse=True)
print(data)
```

```python
cum_count = 0
for k, count in data["R"]:
    print(k, cum_count/(cum_count+count))
    cum_count += count
```
